```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import torch
%matplotlib inline
```

```python
data = pd.read_csv('/DDoSdata.csv')
data['proto'] = data['proto'].str.upper()
data
```

```python
for i in range(len(data)):
  if data['proto'].iloc[i] != data['subcategory'].iloc[i]:
    data.at[i, 'label'] = 1
  else:
    data.at[i, 'label'] = 0
data
```

```python
data['stime'] = data['stime'].astype('category').cat.codes
data['flgs'] = data['flgs'].astype('category').cat.codes
data['proto'] = data['proto'].astype('category').cat.codes
data['saddr'] = data['saddr'].astype('category').cat.codes
data['daddr'] = data['daddr'].astype('category').cat.codes
data['state'] = data['state'].astype('category').cat.codes
data['ltime'] = data['ltime'].astype('category').cat.codes
data['category'] = data['category'].astype('category').cat.codes
data['subcategory'] = data['subcategory'].astype('category').cat.codes
```

```python
x = data.drop('label', axis = 1)
y = data['label']
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy = "median")
imputer.fit(x)
x = imputer.transform(x)
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=41)
```

```python
# Train the model
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
model.fit(x_train, y_train)
```

```python
model.predict(x_test)
```

```python
y_test
```

```python
dict = {1.0: 'malicious', 0.0:'not malicious'}
predictions = model.predict(x)
predictions = np.where(predictions > 0, 'malicious', 'not malicious')
```
